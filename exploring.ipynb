# Kuromoto Definitions

Kuramoto Model:
$$ \frac{d\theta_i}{dt} = \omega_i + \sum^{N}_{j} W_{ij}sin(\theta_j - \theta_i - \alpha) $$

$G$ is a graph with nodes positioned at $(x,y,z) \subset \R^3$ with Adjaceny Matrix $A$ which is non symmetric and real valued. 

The probability for an entry $A_{ij}$ to exist, $P(A_{ij}) = P(A_{ji})$ is a function to be set. 
$W$ is an anti-symmetric weight matrix which is used to scale the impact of an oscillators.
$W_{ij}$ is subject to a Hebbian learning rule which will favour synchronous oscillators.
